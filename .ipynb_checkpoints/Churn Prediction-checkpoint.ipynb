# Telecom Churn Prediction

```python
#Importing packages and loading dataset
import numpy as np
import pandas as pd
import os
```

```python
data=pd.read_csv('F:\\ML Project\\Telco-Customer-Churn.csv')
```

```python
data
```

```python
#creating a copy of data
df=data.copy()
```

## 1.Data preprocessing

```python
data.isnull().sum()
#there is no null values
```

```python
#converting datatype of Totalcharges to numeric dataype
data.TotalCharges = pd.to_numeric(data.TotalCharges, errors='coerce')
data.isnull().sum()
```

```python
#Removing missing values 
data.dropna(how = 'any', inplace = True)
```

```python
data['Churn'] = np.where(data.Churn == 'Yes',1,0)
```

```python
#dropping not required columns
columns_to_drop=['customerID','OnlineBackup','PaperlessBilling']
data.drop(columns=columns_to_drop,axis=1, inplace=True)
data
```

```python
#checking duplicates values and removing it
data.duplicated().sum()
data=data.drop_duplicates(keep='first')
data
```

```python
data.describe()
```

### customer having some extra features will churned or not??
- converting some features to numeric data to analyse the churning of customer.

```python
from sklearn.preprocessing import LabelEncoder

# Apply label encoding to convert categorical variables to numeric values
label_encoder = LabelEncoder()
df['OnlineSecurity'] = label_encoder.fit_transform(df['OnlineSecurity'])
df['DeviceProtection'] = label_encoder.fit_transform(df['DeviceProtection'])
df['TechSupport'] = label_encoder.fit_transform(df['TechSupport'])
df['StreamingTV'] = label_encoder.fit_transform(df['StreamingTV'])
df['StreamingMovies'] = label_encoder.fit_transform(df['StreamingMovies'])
```

```python
different_features=df.groupby('Churn')['OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'].sum()
different_features.plot(kind='bar', stacked=False, figsize=(6,4), color=['skyblue','yellow', 'orange','blue', 'green'])
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.title('Number of Customers by Churn and Feature')
plt.xticks(ticks=[0, 1], labels=['Not Churn', 'Churn'], rotation=0)
plt.legend(title='Feature')
plt.tight_layout()
plt.show()
```

- Customers having online security,device protection,tech support,streaming tv,streaming movies feature are more likely to be not churned

```python
#ploting total charges distribution by internet services
plt.figure(figsize=(6,4))
sns.boxplot(x='InternetService', y='TotalCharges',hue='Churn',data=data)
plt.xlabel('InternetService')
plt.ylabel('Total Charges')
plt.title('Total Charges Distribution by Internet services')
plt.show()
```

- Customer having no internet connection is more likely to be not churned.
- Customer having Fiber optic service can be churned in comparison to dsl.
- Customer having dsl connection are more likely to be not churned.

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

data['tenure_group'] = pd.cut(data.tenure, range(1, 80, 12), right=False, labels=labels)
data['tenure_group'].value_counts()
```

```python
# Calculate churn rate for each tenure group
churn_rate = data.groupby('tenure_group')['Churn'].mean()

plt.figure(figsize=(6,4))
plt.bar(churn_rate.index, churn_rate.values, color='yellow')
plt.xlabel('Tenure Group')
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Tenure Group')
plt.show()
```

- Customer having tenure group 1-12 months are more likely to be churned and the customers having large tenure group are not likely to be churned.

```python
#dropping tenure column
data.drop(columns= ['tenure'], axis=1, inplace=True)
```

```python
#plotting churn rate by contract
different_contracts=data.groupby('Contract')['Churn'].count()
different_contracts.plot(kind='line', title='Churn rate by contract')
plt.xlabel('Contract')
plt.ylabel('Churn')
plt.show()
```

- Customer having contract of month-to-month have high chances for churning

```python
# converting all categorical variable into dummy variable
data_dummy = pd.get_dummies(data)
data_dummy.head()
```

# 3. Splitting into train and test data

```python
#importing libraries to split data
from sklearn.model_selection import train_test_split
X=data_dummy.drop('Churn',axis=1)
Y=data_dummy['Churn']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train.info()
```

## Using Decision tree classifier

```python
#importing libraries for decision trees
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
model=DecisionTreeClassifier(criterion='gini',random_state=100,max_depth=6,min_samples_leaf=8)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
print(classification_report(Y_test,Y_pred,labels=[0,1]))
```

```python
print(confusion_matrix(Y_test,Y_pred))
```

Here accuracy is good but then also 

```python
#Resampling of data because it is highly imbalanced
sm=SMOTEENN()
X_resampled,Y_resampled=sm.fit_resample(X,Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_resampled,Y_resampled,test_size=0.2)
```

```python
new_model=DecisionTreeClassifier(criterion='gini',random_state=100,max_depth=6,min_samples_leaf=8)
```

```python
new_model.fit(x_train,y_train)
```

```python
y_pred=new_model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred,labels=[0,1]))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
new_model.score(x_test,y_test)
```

```python
#importing pickle to save the model
import pickle
pickle.dump(new_model,open('Churn_prediction_model.pkl','wb'))
```
