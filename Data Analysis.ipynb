## Exploratory Analysis and Visualization

```python
# counting weather terms after changing term main to weather
data.rename(columns={'main':'weather'},inplace=True)
data['weather'].value_counts()
```

### Top 10 States with heavy rainfall 

```python
#filtering rain term in weather and then finding top 10 states with heavy rainfall
total_rain=data[data['weather'].str.contains('Rain')]
rain_state=total_rain['state'].value_counts()
rain_state=rain_state[:10]
rain_state
```

```python
plt.figure(figsize=(18,6))
plt.title('Annual Rainfall in states')
plt.bar(rain_state.index,rain_state)
plt.xlabel('State')
plt.ylabel('Rainfall')
plt.show()
```

#####                                                                          Uttar Pradesh has highest rainfall among all states.

### Top 10 Cities having heavy rainfall

```python
city=total_rain['city_name'].value_counts(ascending=False)
city=city[:10]
plt.figure(figsize=(18,6))
plt.title('Annual Rainfall in Cities')
plt.bar(city.index,city, color = 'Blue')
plt.xlabel('City')
plt.ylabel('Rainfall')
plt.show()
```

#### Longleng city has heaviest rainfall among all the other cities

### Analysis of average humidity in different weathers

```python
#Average humidity throughout year in different weathers
different_weathers=data.groupby('weather')['humidity'].mean()
different_weathers.plot(kind='line', title='Average Humidity throughout year in different Weathers')
plt.xlabel('Weathers')
plt.ylabel('Avg Humidity')
plt.show()
```

#### In snow region there is high level of humidity.

### Analysis of average wind speed in different weathers

```python
#Average wind speed throughout year in different weathers
different=data.groupby('weather')['wind_speed'].mean()
different.plot(kind='line', title='Average wind speed throughout year in different weathers')
plt.xlabel('Weathers')
plt.ylabel('Avg wind speed')
plt.show()
```

#### In rainy weather wind blows faster

### Analysis of Cities having maximum temperature

```python
#Cities having maximum temperature
City=data.groupby('city_name')['temp_max'].max()
City=City[:10]
plt.figure(figsize=(18,6))
plt.title('Cities having maximum temperature in Summer')
plt.bar(City.index,City, color = 'Orange')
plt.xlabel('City')
plt.ylabel('Temperaure')
plt.show()
```
